```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs

def parse(driver , page):
    data = {"name":[],"player_value":[]}
    url = f'https://www.transfermarkt.com/premier-league/marktwerte/wettbewerb/GB1/ajax/yw1/page/{page}'
    
    driver.get(url)
    soup = bs(driver.page_source , "html.parser")#.find('div' , class_ = 'responsive_table')
        
    all_data = []
    for soupie in soup.find_all('td',class_ = 'hauptlink'):
        try:
            print(all_data)
            all_data.append(soupie.find('a').string)
        except:
            continue
            
    data["name"]=[all_data[i] for i in range(len(all_data)) if not i%2][:-3]
    data["player_value"]= [all_data[i] for i in range(len(all_data)) if i%2][:-2]


    return data


def parsepages(driver,  num_pages):
    data = {"name":[],"player_value":[]}
    for page in range(1,num_pages+1):
        page_data=parse(driver , page)
        for key in data.keys():
            data[key]+=page_data[key]
    return data

driver = webdriver.Chrome()
all_data = parsepages(driver , 4)
all_data
```

```python
def scrape_all_data(driver):
    seasons = [79,210,274,363,418]
    base_url = "https://www.premierleague.com/stats/top/clubs/wins"
    data = {"rank" : [] ,
                "team" : [] ,
                "season" : []
           }
    
    for se in seasons:
        url = base_url + f"?se={se}"
        driver.get(url)
        time.sleep(5)
        html_content = driver.page_source
        soup = BeautifulSoup(html_content , "html.parser").find("tbody", 
                                                                class_ = "statsTableContainer")

        for soupie in soup.find_all("tr" , class_ = ""):
            rank = soupie.find("td").string

            data["rank"].append(rank.strip())
            team = soupie.find("a" , class_ = "playerName")

            data["team"].append(team.text.strip())
            data["season"].append(se)
    return data

driver = webdriver.Chrome()
driver.get("https://www.premierleague.com/stats/top/clubs/wins?se=418")
time.sleep(5)
data = scrape_all_data(driver)
```
